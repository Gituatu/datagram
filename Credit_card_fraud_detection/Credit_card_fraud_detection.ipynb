```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_c_data= pd.read_csv('/content/creditcard.csv')
```

```python
# Distribution of legit transactions & fraudulent transactions
credit_c_data['Class'].value_counts()
```

This data set is highly unbalanced,
0---> Normal transaction
1---> Fraudulent transaction

```python
# Separating the data for analysis
legit= credit_c_data[credit_c_data.Class==0]
fraud= credit_c_data[credit_c_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# Compare the values for both transactions
credit_c_data.groupby('Class').mean()
```

Under-sampling---> Build a sample dataset containing similer distribution of normal transactions and fraud transactions

Number of fraudulent transactions and normal transactions == 492

```python
legit_sample= legit.sample(n=492)
```

```python
# Concatenating two dataFrames
new_dataset= pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
# Splitting the data into Features & Targets
X= new_dataset.drop(columns='Class', axis=1)
Y= new_dataset['Class']
```

```python
print(X)
print(Y)
```

```python
# Split the data into Training data & Teat data
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Model training using Logistic Regression
model= LogisticRegression()
```

```python
# Training the Logistic Regression Model with the training data
model.fit(X_train, Y_train)
```

```python
# Model evaluation

# Accuracy on training data
X_train_pred= model.predict(X_train)
training_data_accuracy= accuracy_score(X_train_pred, Y_train)
print("Accuracy on training data: ", training_data_accuracy)
```

```python
# Accuracy on test data
X_test_pred= model.predict(X_test)
test_data_accuracy= accuracy_score(X_test_pred, Y_test)
print("Accuracy on test data: ", test_data_accuracy)
```
